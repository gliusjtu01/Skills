```python
import pandas as pd

# Define file path
file_path = "C:/Users/gangl/Documents/Python Scripts/UNCC_Tuition_2000TO2022.csv"

# Load the CSV file
df = pd.read_csv(file_path)

# Display the first few rows to understand the structure
df.head()

import matplotlib.pyplot as plt

# Convert FILE_DATE to a more readable format (extracting the starting year)
df["Year"] = df["FILE_DATE"].str[:4].astype(int)

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(df["Year"], df["TUITIONFEE_IN"], marker="o", linestyle="-", color="b")

# Customize the plot
plt.xlabel("Year")
plt.ylabel("Tuition Fee (In-State) in Dollars")
plt.title("UNC Charlotte Tuition and Fees (2000-2022)")
plt.grid(True)

# Show the plot
plt.savefig("UNC Charlotte Tuition and Fees (2000-2022)", dpi=300, bbox_inches="tight")
plt.show()
```
