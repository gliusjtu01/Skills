```python
import pandas as pd
# Load the new CSV file
file_path = "C:/Users/gangl/Documents/Python Scripts/NC_HighestDegree.csv"
df_degree = pd.read_csv(file_path)

# Display the first few rows to understand the structure
df_degree.head()

import matplotlib.pyplot as plt

# Count occurrences of each degree type
degree_counts = df_degree["HIGHDEG"].value_counts()

# Define labels based on common degree codes (assuming standard categories)
degree_labels = {
    0: "Non-degree-granting",
    1: "Certificate",
    2: "Associate's",
    3: "Bachelor's",
    4: "Graduate"
}

# Map numeric codes to labels
degree_counts.index = degree_counts.index.map(degree_labels)

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(degree_counts, labels=degree_counts.index, autopct='%1.1f%%', colors=["lightblue", "purple","lightgreen", "orange", "red"])
plt.title("Distribution of Highest Degree Awarded in Colleges at North Carolina")

# Show the pie chart
plt.savefig("Distribution of Highest Degree Awarded in Colleges at North Carolina", dpi=300, bbox_inches="tight")
plt.show()
```
