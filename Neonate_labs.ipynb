```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Create a DataFrame from the provided data ---
data = {
    'Labs': ['CBC Obtained', 'LFTs obtained', 'CSF cell counts obtained', 'CSF glucose obtained', 'CSF protein obtained'],
    'Frequency': [95, 9, 30, 30, 30],
    'Percentage': [97.94, 9.28, 30.93, 30.93, 30.93]
}
df = pd.DataFrame(data)

# --- Create the horizontal bar plot ---
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability

# Plot the percentages as bars
plt.barh(df['Labs'], df['Percentage'], color='skyblue')

# Add the frequency values as text labels next to the bars
for index, value in enumerate(df['Percentage']):
    plt.text(value + 2, index, f'({df["Frequency"][index]}) {value:.2f}%', va='center')

# Set labels and title
plt.xlabel('Percentage')
plt.ylabel('Lab Tests')
plt.title('Lab Test Percentage and Frequency')

# Invert the y-axis to display labs in the original order
plt.gca().invert_yaxis()

# Add gridlines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Create a DataFrame from the provided data ---
data = {
    'Labs': ['CBC Obtained', 'LFTs obtained', 'CSF cell counts obtained', 'CSF glucose obtained', 'CSF protein obtained'],
    'Frequency': [95, 9, 30, 30, 30],
    'Percentage': [97.94, 9.28, 30.93, 30.93, 30.93]
}
df = pd.DataFrame(data)

# --- Create the horizontal bar plot ---
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability

# Plot the percentages as bars
plt.barh(df['Labs'], df['Percentage'], color='blue')

# Add the frequency values as text labels next to the bars
for index, value in enumerate(df['Percentage']):
    plt.text(value + 2, index, f'({df["Frequency"][index]}) {value:.2f}%', va='center')

# Set labels and title
plt.xlabel('Percentage (%)', fontweight = 'bold')
plt.title('Proportion of Neonates with Labs Obtained', fontweight = 'bold')

# Set the x-axis limits from 0 to 119
plt.xlim(0, 115)

# Invert the y-axis to display labs in the original order
plt.gca().invert_yaxis()

# Add gridlines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.savefig('Neonates_Lab.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create the data
data = {
    "Labs": [
        "CSF obtained before antibiotics",
        "CSF obtained after antibiotics",
        "Blood Culture obtained before antibiotics",
        "Blood Culture obtained after antibiotics",
        "Urine Culture obtained before antibiotics",
        "Urine Culture obtained after antibiotics"
    ],
    "Frequency": [0, 29, 2, 62, 3, 27],
    "Percentage": [0.00, 100.00, 3.13, 96.88, 10.00, 90.00]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(df["Labs"], df["Percentage"], color="skyblue")

# Add text labels (frequency and percentage)
for bar, freq, perc in zip(bars, df["Frequency"], df["Percentage"]):
    width = bar.get_width()
    label = f"{freq} ({perc:.2f}%)"
    ax.text(width + 1, bar.get_y() + bar.get_height()/2, label, va='center')

# Axis labels and title
ax.set_xlabel("Percentage (%)")
ax.set_title("Lab Timing Relative to Antibiotic Administration")
ax.set_xlim(0, 110)  # give space for labels

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.patches import Patch

# Create the data
data = {
    "Labs": [
        "CSF obtained before antibiotics",
        "CSF obtained after antibiotics",
        "Blood Culture obtained before antibiotics",
        "Blood Culture obtained after antibiotics",
        "Urine Culture obtained before antibiotics",
        "Urine Culture obtained after antibiotics"
    ],
    "Frequency": [0, 29, 2, 62, 3, 27],
    "Percentage": [0.00, 100.00, 3.13, 96.88, 10.00, 90.00],
    "Type": ["CSF", "CSF", "Blood", "Blood", "Urine", "Urine"]
}

# Create DataFrame and reverse for top-to-bottom ordering
df = pd.DataFrame(data)[::-1]

# Assign new color map
color_map = {
    "CSF": "blue",
    "Blood": "orange",
    "Urine": "green"
}
colors = df["Type"].map(color_map)

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(df["Labs"], df["Percentage"], color=colors)

# Add text labels (Frequency and %)
for bar, freq, perc in zip(bars, df["Frequency"], df["Percentage"]):
    width = bar.get_width()
    label = f"{freq} ({perc:.2f}%)"
    ax.text(width + 1.5, bar.get_y() + bar.get_height() / 2, label, va='center')

# Custom legend
legend_elements = [
    Patch(facecolor="blue", label="CSF"),
    Patch(facecolor="orange", label="Blood Culture"),
    Patch(facecolor="green", label="Urine Culture")
]
ax.legend(handles=legend_elements, title="Lab Type", loc="upper right")

# Customize axes and title
ax.set_xlim(0, 119)
ax.set_xlabel("Percentage of Neonates (%)", fontweight='bold')
ax.set_title("Proportion of Neonates with Cultures Obtained", fontweight='bold')

plt.tight_layout()
plt.savefig('Neonates_Lab_figure2.png', dpi=300, bbox_inches='tight')
plt.show()
```
