```python
import pandas as pd

file_path = "C:/Users/gangl/Documents/Python Scripts/TripleCombination_Data_needed.csv"
df = pd.read_csv(file_path)

df
```

```python
# Pivot the data
df_pivoted = df.pivot(index='record_id', columns='redcap_repeat_instance')

df_pivoted
```

```python
# Flatten the MultiIndex columns
df_pivoted.columns = [f'{col}_{inst}' for col, inst in df_pivoted.columns]

df_pivoted
```

```python
# Reset index if needed
df_pivoted.reset_index(inplace=True)

# Display result
df_pivoted
```

```python
df_pivoted.to_csv("Triple.csv", index=False)
```
