```python
import matplotlib.pyplot as plt

# Data for Jiangsu and Zhejiang
labels = ["Agriculture", "Industry & Construction", "Services"]
jiangsu_gdp = [5245.2, 59180.1, 72582.8]  # in billion RMB
zhejiang_gdp = [2586, 34800, 52800]  # in billion RMB

# Total GDP for scaling pie chart sizes
jiangsu_total = sum(jiangsu_gdp)
zhejiang_total = sum(zhejiang_gdp)

# Pie chart sizes proportional to GDP
sizes = [jiangsu_total / 1000, zhejiang_total / 1000]  # Scaling down for visualization

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Colors for consistency
colors = ["#ff9999","#66b3ff","#99ff99"]

# Plot Jiangsu
axes[0].pie(jiangsu_gdp, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, radius=1.2)
axes[0].set_title(f"Jiangsu GDP Composition (Total: {jiangsu_total:.1f}B RMB)")

# Plot Zhejiang
axes[1].pie(zhejiang_gdp, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, radius=zhejiang_total/jiangsu_total * 1.2)
axes[1].set_title(f"Zhejiang GDP Composition (Total: {zhejiang_total:.1f}B RMB)")

plt.suptitle("Comparison of 2024 GDP Composition: Jiangsu vs Zhejiang", fontsize=14)
plt.savefig('Comparison of 2024 GDP Composition: Jiangsu vs Zhejiang.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "SimHei"  # 黑体字体，支持中文
plt.rcParams["axes.unicode_minus"] = False  # 解决负号显示问题

# Data for Jiangsu and Zhejiang
labels = ["农业", "工业和建筑业", "服务业"]
jiangsu_gdp = [5245.2, 59180.1, 72582.8]  # in billion RMB
zhejiang_gdp = [2586, 34800, 52800]  # in billion RMB

# Total GDP for scaling pie chart sizes
jiangsu_total = sum(jiangsu_gdp)
zhejiang_total = sum(zhejiang_gdp)

# Pie chart sizes proportional to GDP
sizes = [jiangsu_total / 1000, zhejiang_total / 1000]  # Scaling down for visualization

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Colors for consistency
colors = ["#ff9999","#66b3ff","#99ff99"]

# Plot Jiangsu
axes[0].pie(jiangsu_gdp, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, radius=1.2)
axes[0].set_title(f"江苏GDP构成 (总计13.7万亿元；人均16.1万元)")

# Plot Zhejiang
axes[1].pie(zhejiang_gdp, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, radius=13.6/16.1 * 1.2)
axes[1].set_title(f"浙江GDP构成 (总计：9.02万亿元；人均13.6万元)")

plt.text(0, 0.5, "数据大观", fontsize=40, color="gray",
         alpha=0.3, ha='center', va='center', transform=plt.gca().transAxes, rotation=30)

plt.suptitle("2024年江苏省和浙江省GDP构成的对比", fontsize=14)
plt.savefig('2024年江苏省和浙江省GDP构成的对比.png')
plt.show()
```
