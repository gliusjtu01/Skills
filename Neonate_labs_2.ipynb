```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Create a DataFrame from the provided data ---
data = {
    'Labs': ['CBC Obtained', 'LFTs obtained', 'CSF cell counts obtained', 'CSF glucose obtained', 'CSF protein obtained'],
    'Frequency': [95, 9, 30, 30, 30],
    'Percentage': [97.94, 9.28, 30.93, 30.93, 30.93]
}
df = pd.DataFrame(data)

# --- Create the horizontal bar plot ---
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability

# Plot the percentages as bars
plt.barh(df['Labs'], df['Percentage'], color='skyblue')

# Add the frequency values as text labels next to the bars
for index, value in enumerate(df['Percentage']):
    plt.text(value + 2, index, f'({df["Frequency"][index]}) {value:.2f}%', va='center')

# Set labels and title
plt.xlabel('Percentage')
plt.ylabel('Lab Tests')
plt.title('Lab Test Percentage and Frequency')

# Invert the y-axis to display labs in the original order
plt.gca().invert_yaxis()

# Add gridlines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Create a DataFrame from the provided data ---
data = {
    'Labs': ['CBC Obtained', 'LFTs obtained', 'CSF cell counts obtained', 'CSF glucose obtained', 'CSF protein obtained'],
    'Frequency': [95, 9, 30, 30, 30],
    'Percentage': [97.94, 9.28, 30.93, 30.93, 30.93]
}
df = pd.DataFrame(data)

# --- Create the horizontal bar plot ---
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability

# Plot the percentages as bars
plt.barh(df['Labs'], df['Percentage'], color='blue')

# Add the frequency values as text labels next to the bars
for index, value in enumerate(df['Percentage']):
    plt.text(value + 2, index, f'({df["Frequency"][index]}) {value:.2f}%', va='center')

# Set labels and title
plt.xlabel('Percentage (%)', fontweight = 'bold')
plt.title('Proportion of Neonates with Labs Obtained', fontweight = 'bold')

# Set the x-axis limits from 0 to 119
plt.xlim(0, 115)

# Invert the y-axis to display labs in the original order
plt.gca().invert_yaxis()

# Add gridlines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.savefig('Neonates_Lab.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create the data
data = {
    "Labs": [
        "CSF obtained before antibiotics",
        "CSF obtained after antibiotics",
        "Blood Culture obtained before antibiotics",
        "Blood Culture obtained after antibiotics",
        "Urine Culture obtained before antibiotics",
        "Urine Culture obtained after antibiotics"
    ],
    "Frequency": [0, 29, 2, 62, 3, 27],
    "Percentage": [0.00, 100.00, 3.13, 96.88, 10.00, 90.00]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(df["Labs"], df["Percentage"], color="skyblue")

# Add text labels (frequency and percentage)
for bar, freq, perc in zip(bars, df["Frequency"], df["Percentage"]):
    width = bar.get_width()
    label = f"{freq} ({perc:.2f}%)"
    ax.text(width + 1, bar.get_y() + bar.get_height()/2, label, va='center')

# Axis labels and title
ax.set_xlabel("Percentage (%)")
ax.set_title("Lab Timing Relative to Antibiotic Administration")
ax.set_xlim(0, 110)  # give space for labels

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.patches import Patch

# Create the data
data = {
    "Labs": [
        "CSF obtained before antibiotics",
        "CSF obtained after antibiotics",
        "Blood Culture obtained before antibiotics",
        "Blood Culture obtained after antibiotics",
        "Urine Culture obtained before antibiotics",
        "Urine Culture obtained after antibiotics"
    ],
    "Frequency": [0, 29, 2, 62, 3, 27],
    "Percentage": [0.00, 100.00, 3.13, 96.88, 10.00, 90.00],
    "Type": ["CSF", "CSF", "Blood", "Blood", "Urine", "Urine"]
}

# Create DataFrame and reverse for top-to-bottom ordering
df = pd.DataFrame(data)[::-1]

# Assign new color map
color_map = {
    "CSF": "blue",
    "Blood": "orange",
    "Urine": "green"
}
colors = df["Type"].map(color_map)

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(df["Labs"], df["Percentage"], color=colors)

# Add text labels (Frequency and %)
for bar, freq, perc in zip(bars, df["Frequency"], df["Percentage"]):
    width = bar.get_width()
    label = f"{freq} ({perc:.2f}%)"
    ax.text(width + 1.5, bar.get_y() + bar.get_height() / 2, label, va='center')

# Custom legend
legend_elements = [
    Patch(facecolor="blue", label="CSF"),
    Patch(facecolor="orange", label="Blood Culture"),
    Patch(facecolor="green", label="Urine Culture")
]
ax.legend(handles=legend_elements, title="Lab Type", loc="upper right")

# Customize axes and title
ax.set_xlim(0, 119)
ax.set_xlabel("Percentage of Neonates (%)", fontweight='bold')
ax.set_title("Proportion of Neonates with Cultures Obtained", fontweight='bold')

plt.tight_layout()
plt.savefig('Neonates_Lab_figure2.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# TableB data
data = {
    "Labs": ["CSF", "CSF", "CSF", "Blood Culture", "Blood Culture", "Blood Culture", "Urine Culture", "Urine Culture", "Urine Culture"],
    "Category": [
        "Obtained before antibiotics", "Obtained after antibiotics", "Unknown",
        "Obtained before antibiotics", "Obtained after antibiotics", "Unknown",
        "Obtained before antibiotics", "Obtained after antibiotics", "Unknown"
    ],
    "Percentage": [0.0, 96.7, 3.3, 2.7, 84.9, 12.3, 9.4, 84.4, 6.3]
}

df = pd.DataFrame(data)

# Combine Labs and Category for Y-axis labels
df["Label"] = df["Labs"] + " - " + df["Category"]

# Color mapping for Lab types
color_map = {
    "CSF": "steelblue",
    "Blood Culture": "orange",
    "Urine Culture": "green"
}
df["Color"] = df["Labs"].map(color_map)

# Plotting horizontal bar plot
fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.barh(df["Label"], df["Percentage"], color=df["Color"])

# Add percentage text labels on the bars
for bar, perc in zip(bars, df["Percentage"]):
    ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f"{perc:.1f}%", va='center')

# Custom legend
legend_elements = [
    Patch(facecolor="steelblue", label="CSF"),
    Patch(facecolor="orange", label="Blood Culture"),
    Patch(facecolor="green", label="Urine Culture")
]
ax.legend(handles=legend_elements, title="Lab Type", loc="lower right")

# Axis labels and title
ax.set_xlabel("Percentage of Neonates (%)")
ax.set_title("TableB: Lab Timing Relative to Antibiotics (Neonates)", fontsize=14, fontweight="bold")
ax.set_xlim(0, 110)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

# TableB data
data = {
    "Labs": ["CSF", "CSF", "CSF", "Blood Culture", "Blood Culture", "Blood Culture", "Urine Culture", "Urine Culture", "Urine Culture"],
    "Category": [
        "Before antibiotics", "After antibiotics", "Unknown",
        "Before antibiotics", "After antibiotics", "Unknown",
        "Before antibiotics", "After antibiotics", "Unknown"
    ],
    "Percentage": [0.0, 96.7, 3.3, 2.7, 84.9, 12.3, 9.4, 84.4, 6.3]
}

df = pd.DataFrame(data)

# Define bar colors by Category
category_colors = {
    "Before antibiotics": "skyblue",
    "After antibiotics": "orange",
    "Unknown": "gray"
}
df["Color"] = df["Category"].map(category_colors)

# Unique labs and categories
labs = ["CSF", "Blood Culture", "Urine Culture"]
categories = ["Before antibiotics", "After antibiotics", "Unknown"]

# Bar positioning offsets
bar_height = 0.2
y_positions = np.arange(len(labs))

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars for each category
for i, category in enumerate(categories):
    subset = df[df["Category"] == category]
    ax.barh(y_positions + (i - 1) * bar_height, subset["Percentage"], height=bar_height, color=category_colors[category], label=category)

# Y-axis ticks for Labs only
ax.set_yticks(y_positions)
ax.set_yticklabels(labs)

# Labels, limits, title
ax.set_xlabel("Percentage of Neonates (%)")
ax.set_xlim(0, 110)
ax.set_title("Lab Timing Relative to Antibiotics (TableB)", fontsize=14, fontweight="bold")

# Legend for Categories
ax.legend(title="Category", loc="lower right")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# TableB data
data = {
    "Labs": ["CSF", "CSF", "CSF", "Blood Culture", "Blood Culture", "Blood Culture", "Urine Culture", "Urine Culture", "Urine Culture"],
    "Category": [
        "Before antibiotics", "After antibiotics", "Unknown",
        "Before antibiotics", "After antibiotics", "Unknown",
        "Before antibiotics", "After antibiotics", "Unknown"
    ],
    "Percentage": [0.0, 96.7, 3.3, 2.7, 84.9, 12.3, 9.4, 84.4, 6.3]
}

df = pd.DataFrame(data)

# Define color for each Category
category_order = ["Before antibiotics", "After antibiotics", "Unknown"]
category_colors = {
    "Before antibiotics": "blue",
    "After antibiotics": "orange",
    "Unknown": "green"
}

# Lab order (CSF on top)
labs_order = ["Urine Culture", "Blood Culture", "CSF"]

# Bar plot parameters
bar_height = 0.2
y_base = np.arange(len(labs_order))

# Plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot grouped bars in specified top-to-bottom category order
for i, category in enumerate(category_order):
    subset = df[df["Category"] == category]
    subset = subset.set_index("Labs").loc[labs_order].reset_index()  # Ensure correct Lab order
    ax.barh(y_base + (1 - i) * bar_height, subset["Percentage"], height=bar_height, color=category_colors[category], label=category)

# Set Y-axis labels for Labs
ax.set_yticks(y_base)
ax.set_yticklabels(labs_order)

# Titles and labels
ax.set_xlabel("Percentage of Patients (%)")
ax.set_xlim(0, 110)
ax.set_title("Proportion of Neonates with Cultures Obtained", fontsize=14, fontweight="bold")

# Legend by Category
ax.legend(title="Category", loc="lower right")
plt.savefig('Figure2.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
